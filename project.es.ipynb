#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
indice_max = df['price'].idxmax()
casa_mas_cara = df.loc[indice_max]
print(f"La casa con direccion en calle: {casa_mas_cara['address']}, es la mas cara y su precio es de €{casa_mas_cara['price']}")
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
indice_min = df['price'].idxmin()
casa_mas_barata = df.loc[indice_min]
print(f"La casa con direccion en calle: {casa_mas_barata['address']}, es la mas barata y su precio es de €{casa_mas_barata['price']}")
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt
poblacion_objetivo = 'Arroyomolinos (Madrid)'
ds_filtrado = ds[ds['level5'] == poblacion_objetivo].copy()
prices = ds_filtrado[ds_filtrado['price'] > 0]['price']
plt.figure(figsize=(10, 6))
plt.hist(prices, bins=20, edgecolor='black', color='skyblue')
plt.title(f'Histograma de Precios para {poblacion_objetivo}', fontsize=14)
plt.xlabel('Precio (€)', fontsize=12)
plt.ylabel('Frecuencia', fontsize=12)
plt.grid(axis='y', alpha=0.7)
plt.show()
```

En el histograma de precios de Arroyomolinos (Madrid), se puede observar que el precio promedio o de la  zona ronda entre los 300000 y 400000, aunque tambien se puede ver un rango bastante amplio entre los precios minimos y maximos, algo normal en el mercado inmobiliario.
